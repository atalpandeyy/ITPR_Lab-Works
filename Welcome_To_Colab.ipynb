```
def calculate_simple_interest(principal, rate, time):
    """
    Calculates the simple interest.

    Args:
        principal (float): The principal amount.
        rate (float): The annual interest rate (as a percentage).
        time (float): The time in years.

    Returns:
        float: The calculated simple interest.
    """
    # Convert rate from percentage to decimal
    rate_decimal = rate / 100
    simple_interest = principal * rate_decimal * time
    return simple_interest

# Example usage:
principal_amount = 1000
interest_rate = 5   # 5%
time_period = 3     # 3 years

interest = calculate_simple_interest(principal_amount, interest_rate, time_period)
print(f"Principal Amount: ${principal_amount:.2f}")
print(f"Interest Rate: {interest_rate}%")
print(f"Time Period: {time_period} years")
print(f"Simple Interest: ${interest:.2f}")
```
